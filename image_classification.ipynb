```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Path ke dataset
dataset_path = 'dataset/anemic'  # ganti sesuai folder kamu

image_files = [f for f in os.listdir(dataset_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
print(f"Jumlah total gambar: {len(image_files)}")

print("List File: ")
image_files[:10]
```

```python
for img_file in image_files[:3]:
    img_path = os.path.join(dataset_path, img_file)
    img = Image.open(img_path)
    plt.imshow(img)
    plt.title(f"{img_file} | Ukuran: {img.size}")
    plt.axis('off')
    plt.show()
```
